# Read a new story


```python
with open("news_story.txt","r") as f :
    news_text = f.read()
news_text[:500]
```

```python
import spacy 

nlp = spacy.load("en_core_web_sm")


doc = nlp(news_text)

numeral_token = []
noun_token = []

for token in doc:
    if token.pos_ =='NOUN':
        noun_token.append(token)
    elif token.pos_ =='NUM':
        numeral_token.append(token)

# numeral_token[:10]
noun_token[:10]
```

```python
count =  doc.count_by(spacy.attrs.POS)
count
```

```python
for k ,v in count.items():
    # print(count.items)
    print(doc.vocab[k].text, "|",v)
```
